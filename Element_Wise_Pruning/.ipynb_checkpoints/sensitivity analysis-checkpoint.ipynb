```python
import os
import sys
import re
from collections import OrderedDict
```

```python
def sensitivity_analysis(fname):

    dictionary_top1 = OrderedDict()
    dictionary_top5 = OrderedDict()
    dictionary_loss = OrderedDict()

    with open(fname) as fp:  
        for cnt, line in enumerate(fp):
            #print("Line {}: {}".format(cnt, line))
            pattern1 = re.compile(r"(\S+) (\S+) - Testing sensitivity of (\S+) \[(\d+.\d+)% sparsity\]")
            match1 = pattern1.match(line)
            if match1:
                ser_date, ser_no, model_name, sparsity_str = match1.groups()
                #print('{:s}, {:s}'.format(model_name, sparsity_str))
            
            pattern2 = re.compile(r"(\S+) (\S+) - ==> Top1: (\d+.\d+)    Top5: (\d+.\d+)    Loss: (\d+.\d+)")
            match2 = pattern2.match(line)
            if match2:
                ser_date, ser_no, top1_str, top5_str, loss_str = match2.groups()
                #print('{:s}, {:s}, {:s}, {:s}, {:s}'.format(model_name, sparsity_str, top1_str, top5_str, loss_str))
                sparsity = float(sparsity_str)
                top1 = float(top1_str)
                top5 = float(top5_str)
                loss = float(loss_str)
            
                if (model_name in dictionary_top1) == False:
                    dictionary_top1[model_name] = OrderedDict()
                if (model_name in dictionary_top5) == False:
                    dictionary_top5[model_name] = OrderedDict()
                if (model_name in dictionary_loss) == False:
                    dictionary_loss[model_name] = OrderedDict()
            
                #print(float(top1))
                dictionary_top1[model_name][sparsity] = top1
                dictionary_top5[model_name][sparsity] = top5
                dictionary_loss[model_name][sparsity] = loss
            
    #print(dictionary_top1)
    return dictionary_top1, dictionary_top5, dictionary_loss
```

```python
import matplotlib.pyplot as plt

def show_sensitivity(dict_acc, sense_type):
    plt.figure(figsize=(20,20))

    for mo in dict_acc:
    #for mo in ['module.conv1.weight']:
        plt.plot(list(dict_acc[mo].keys()), list(dict_acc[mo].values()))

    plt.xticks(fontsize=20)
    plt.yticks(fontsize=20)

    plt.xlabel('Sparsity', fontsize=20)
    plt.ylabel('Accuracy', fontsize=20)
    plt.title(sense_type + ' Sensitivity Analysis on Resnet20_Cifar', fontsize=20) 

    plt.legend(list(iter(dict_acc)), fontsize=20)

    plt.grid()
    plt.show()
```

```python
fname = './resnet20_cifar element sensitivity anaysis.log'
dictionary_top1, dictionary_top5, dictionary_loss = sensitivity_analysis(fname)
print(dictionary_top1)
```

```python
show_sensitivity(dictionary_top1, "Element")
```
